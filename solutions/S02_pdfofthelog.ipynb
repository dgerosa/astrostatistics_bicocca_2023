```python
import numpy as np
import pylab as plt
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
x=np.random.uniform(0.1,10,10000)
```

```python
x
```

```python
plt.hist(x,density=True,histtype='step',lw=2);
plt.xlabel("$x$");
```

```python
y=np.log10(x)
plt.hist(y,density=True,histtype='step',lw=2);
plt.xlabel("$y$");
```

Compute the pdf

$y=\log_{10} x$

$x=10^y$

$\frac{dx}{dy} = 10^y \log(10)$

Uniform in x:
$p(x) = \frac{1}{10-0.1}$


Therefore $p(y) =10^y \log(10) /(10-0.1)$



```python
ygrid=np.linspace(-1,1,100)
pofy = 10**ygrid *np.log(10)/(10-0.1)

plt.plot(ygrid,pofy);
```

```python
plt.hist(y,density=True,histtype='step',lw=2);
plt.plot(ygrid,pofy);
plt.xlabel("$y$");
```
